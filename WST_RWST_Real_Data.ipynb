```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from os import listdir
import os
import sys
sys.path.insert(0, os.path.abspath('..')) # Note that this line is useless with a regular pip installation of PyWST.
import pywst as pw
```

```python
ddir = '/Users/ianhothi/WST/Testing/pywst/examples/data/'
```

```python
Stokes_I = fits.getdata(ddir+'C007010_Ibt_fov4_u50-250.fits')
Stokes_V = fits.getdata(ddir+'C007010_Vbt_fov4_u50-250.fits')
```

```python
plt.figure(figsize=(6,6))
plt.imshow(Stokes_I[0,0])
```

```python
plt.figure(figsize=(6,6))
plt.imshow(Stokes_V[0,0])
```

```python
M, N = 256,256 # Height and width of the test maps
J = 8 # We choose to probe 7 scales
L = 8 # Number of angles in which the interval [0,pi] is divided
OS = 0 # No oversampling
```

```python
rwst_op = pw.RWSTOp(M, N, J, L, OS)
```

```python
wst_op = pw.WSTOp(M, N, J, L, OS)
```

```python
%matplotlib inline
```

```python
Stokes_I_WST = wst_op.apply(Stokes_I[0,:,128:256+128,128:256+128])
Stokes_I_WST.average()
Stokes_I_WST.normalize()
```

```python
Stokes_I_WST.plot()
```

```python
Stokes_I_RWST = rwst_op.apply(Stokes_I[0,:,128:256+128,128:256+128],wst_average=True)
```

```python
Stokes_I_RWST.plot()
```
