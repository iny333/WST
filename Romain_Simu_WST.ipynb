```python
def read_cube(path, type = np.float):
    try :
        cube = np.fromfile(path, dtype=type)
    except FileNotFoundError :
        print(" !!!!!! file not found : "+ path)
        cube = np.zeros(256**3)
        print("moving on...")
    #print('cube loaded')
    shape = np.shape(cube)[0]
    #print( 'shape = ', shape)
    length = int(shape**(1/3)) +1

    cube =np.reshape(cube, (length,length,length)).T
    shape = np.shape(cube)
    #print('av, min, max : ',np.mean(cube), np.min(cube), np.max(cube))
    return cube
```

I chose to vary one parameter for star formation (Mmin), one for x-ray emissivity (fx), one for UV emissivit(escape fraction fesc_post)

Cubes of physical quantities are in simu000XX/postprocessing/cubes

They are 256*256*256 cubes. example.py shows one way of reading them
simu000XX/postprocessing/z_dtb_tneutre_xa.txt has four columns of global quantities : redshift, global brightness temperature, average temperature of the neutral gas, lyman-alpha coupling
These are the redshifts at which the cubes were computed.
